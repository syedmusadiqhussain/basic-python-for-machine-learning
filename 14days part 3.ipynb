# Matplotlib library

```python
import matplotlib.pyplot as plt
```

# linear graph

```python
x=[2,3,4,5]
y=[6,7,8,9]
c= 'r'
plt.plot(x,y,c)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,3,4, 24]  # Adjusted length
y = [3.2,1,3,33]  # Same length as x

plt.plot(x, y,marker='o')
plt.show()
```

```python
x=[3,5,3,2,3]
y=[9,3,2,8,5]
plt.plot(x,y)
plt.plot(x,marker='>')
plt.show()
```

# bar graph

```python
x=[2,3,4,4]
y=[2,34,42,2]
plt.bar(x,y)
plt.show()
```

# color to bar graph

```python
x=[2,3,4,4]
y=[2,34,42,2]
c=['y','g','r','b']
plt.bar(x,y,color=c)
plt.show()
```
