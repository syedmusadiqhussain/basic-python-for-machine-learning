# Seaborn

```python
import pandas as pd
var = [1, 2, 3]
var1 = [4, 5, 6]
df = pd.DataFrame({"var": var, "var1": var1})
print(df)
sns.lineplot(x='var',y='var1', data=df)
plt.show() #in seaborn the x_axis and y_axis atomatic data or imported

```

# penguins data

```python
import pandas as pd
df1 = pd.read_csv("penguins.csv")


```

```python
df1
```

```python
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", data=df1)
# Show the plot
plt.show() #seaborn direct x_axis or y_axis
```

```python
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", data=df1, hue="sex")
# in this graph the male and female divided due to hue
plt.show() #seaborn direct x_axis or y_axis
```

```python
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", data=df1, hue="sex", palette={"Male": "blue", "Female": "red"})
plt.show()
```
